### Bar chart
To chart the values 25, 5, 150, and 100 into the bar graph, we will store these balues in an array and pass it to the `bar` function, which is part of the `matplotlib.pyplot` class.

```python
import matplotlib.pyplot as plt

data = [25., 5., 150., 100.]
x_values = range(len(data))
plt.figure(figsize = (12, 8))
plt.bar(x_values, data)
plt.show()
```

We can change the width of the bars by using the `width` argument.

```python
import matplotlib.pyplot as plt

data = [25., 5., 150., 100.]
x_values = range(len(data))
plt.figure(figsize = (12, 8))
plt.bar(x_values, data, width = 1.)
plt.show()
```

However, this is not visually appealing because there is no space between the bars anymore.
### Multiple bar charts
In trying to visualize data, stacking a number of bars enables one to further understand how one piece of data or variable varies compared to another:

```python
import numpy as np

data = [
    [8., 57., 22., 10.],
    [16., 7., 32., 40.]
]

x_values = np.arange(4)

plt.figure(figsize = (12, 8))
plt.bar(x_values + 0., data[0], color = "r", width = 0.3)
plt.bar(x_values + 0.3, data[1], color = "y", width = 0.3)
plt.show()
```

The y values for the first batch of data is the first row of the matrix. The second batch is the second row of the matrix. When the bars are plotted, they will occupy the same x position, side by side.

`x_values = np.arange(4)` generates an array with values `[0, 1, 2, 3]`. The first set of bars are drawn first at position `x_values + 0.30`. Thus, the first x-values will be plotted at 0, 1, 2, and 3. The second batch of `x_values` will be plotted at 0.30, 1.30, 2.30, and 3.30.
### Box plot
The box plot is used to visualize the median value and low and high ranges of a distribution. It is also referred to as a box and whisker plot.

Let's chart a simple box plot. We begin by generating 50 numbers from a normal distribution. These are then passed to `plt.boxplot(data)` to be charted:

```python
data = np.random.randn(50)

plt.figure(figsize = (10, 6))
plt.boxplot(data)
plt.show()
```

A boxplot is useful to easily identify the outliers in a dataset, as well as determining in  which direction a dataset may be skewed.
### Pie chart
The pie chart interprets and visually represents the data as if to fit into a circle. The individual data points are expressed as sectors of a circle that add up to 360 degrees/$2\pi$ radians. This chart is okay for displaying categorical data and summaries too:

```python
data = [5, 2, 2.5]
labels = ["Agriculture", "Aide", "News"]

plt.figure(figsize = (8, 6))
plt.pie(data, labels = labels, autopct = "%1.1f%%")
plt.show()
```

### Bubble chart
Another variant of the scatter plot is the bubble chart. In a scatter plot, we only plot the x and y points of the data. Bubble charts add another dimension by illustrating the size of the points. THe third dimension may represent something else:

```python
n = 10
x = np.random.rand(n)
y = np.random.rand(n)
colors = np.random.rand(n)
area = np.pi * (60 * np.random.rand(n)) ** 2

plt.figure(figsize = (12, 16))
plt.subplot(211)
plt.title("Normal Scatter Plot")
plt.scatter(x, y, c = colors, alpha = 0.75)

plt.subplot(212)
plt.title("Bubble Scatter Plot")
plt.scatter(x, y, s = area, c = colors, alpha = 0.75)
plt.show()
```
